```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
  
  
def delta(theta, n, beta):
    gamma = 1/np.sqrt(1-beta**2)

    return gamma**2* n**2* ((beta**2-1)**2 * gamma**2 * np.cos(theta)**2 + np.sin(theta)**2 * (1-beta**2 * n**2))
```

```python
def vp(theta, n, beta):
    gamma = 1/np.sqrt(1-beta**2)
    return (beta * gamma**2 * (n**2 -1)* np.cos(theta) + np.sqrt(delta(theta, n, beta))) / (gamma**2 * (n**2 - beta**2)*np.cos(theta)**2 + n**2* np.sin(theta)**2)
def vm(theta, n, beta):
    gamma = 1/np.sqrt(1-beta**2)
    return (beta * gamma**2 * (n**2 -1)* np.cos(theta) - np.sqrt(delta(theta, n, beta))) / (gamma**2 * (n**2 - beta**2)*np.cos(theta)**2 + n**2* np.sin(theta)**2)
```

```python
beta = 0.4
n = 1.5
theta = np.linspace(-np.pi, np.pi, 1000000)
plt.rc("text", usetex= True)
plt.plot(theta, vp(theta, n, beta))
plt.plot(theta, vm(theta, n, beta), c="tab:blue")
plt.gca().set_ylim([0, 1])
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ["$-\\pi$", "$-\\pi/2$", '$0$', "$\\pi/2$", '$\\pi$'])
plt.gca().tick_params(axis="both", which="major", labelsize=18)
plt.xlabel("$\\theta'$", fontsize=25)
plt.ylabel("$v'(\\theta')$", fontsize=25)
plt.tight_layout()

plt.savefig("slow-moving-object.pdf", dpi=1000)
```
